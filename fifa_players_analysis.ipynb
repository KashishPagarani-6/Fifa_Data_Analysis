```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
fifa.Weight
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
plt.style.use('ggplot')
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]
weights = [light,light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175- 200', 'Over 200']
plt.pie(weights, labels= labels, autopct = '%.2f%%', pctdistance = 0.8)
plt.show()
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
plt.style.use('ggplot')
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]
weights = [light,light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175- 200', 'Over 200']
explode = (.5,.2,.0,.0,.5)
plt.pie(weights, labels= labels, autopct = '%.2f%%', pctdistance = 0.8, explode = explode)
plt.title('Weight Distribution of FIFA Players (in lbs)')
plt.savefig("weight_distribution_of_fifaPlayers.png") 
plt.show()
```

```python
plt.style.use('default')
Barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
Madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
Juventus = fifa.loc[fifa.Club == 'Juventus']['Overall']
Revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
labels = ['FC Barcelona','Real Madrid','Juventus','NE Revolution']
# plt.boxplot() does not take multiple datasets as separate arguments. It needs one iterable containing all datasets.So we'll add all these in a list
boxes = plt.boxplot([Barcelona, Madrid, Juventus, Revs], tick_labels=labels, patch_artist=True)
for box in boxes['boxes']:
    # Set edge color
    box.set(color = '#4286f4', linewidth =2)
    # Set fill color
    box.set(facecolor = '#e0e0e0')
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.savefig("Soccre_team.png") 
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(fifa['Age'], fifa['Overall'])
plt.xlabel("Age")
plt.ylabel("Overall Rating")
plt.title("Age vs Player Performance")
plt.savefig("age_vs_player.png") 
plt.show()
```

```python
skills = fifa[['Finishing','Dribbling','BallControl','SprintSpeed',
             'Strength','ShotPower','Stamina','Vision','Overall']]

plt.figure(figsize=(10,7))
sns.heatmap(skills.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Player Skills")
plt.savefig("correlation_btw_skills.png") 
plt.show()
```

```python
top10 = fifa.sort_values('Overall', ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.barh(top10['Name'], top10['Overall'])
plt.xlabel("Overall Rating")
plt.title("Top 10 Highest Rated FIFA Players")
plt.gca().invert_yaxis()
plt.savefig("top10players.png") 
plt.show()
```

```python
top_countries = fifa['Nationality'].value_counts().head(10)
plt.figure(figsize=(10,6))
plt.bar(top_countries.index, top_countries.values)
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Number of Players")
plt.title("Top 10 Countries Producing Players")
plt.savefig("topCountries.png") 
plt.show()
```
